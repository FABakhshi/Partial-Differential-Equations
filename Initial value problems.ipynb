Malthusian growth model:

A Malthusian growth model, is a exponential growth according to
a function being proportional to its speed of grows. In fact, Malthusian growth models have the following form:

$P(t)=P_{0}\ e^{kt}$ where, $P(0)$ is the initial population size (population at time $t=0$) and $k$ is the population growth rate. In fact, $P$ satisfies the following differential equation (with an initial condition):
$$\Biggl\{\begin{array}{ll}
    \frac{dP}{dt}(t)=kP(t), & 0<t\leq T \\\\
    P(0)=P_0.  
\end{array}$$

Explicit/Forward Euler Method:

We consider the following initial value problem:
$$\Biggl\{\begin{array}{ll}
    y'(t)=f(y(t)), & 0<t\leq T \\\\
    y(0)=y_0.  
\end{array}$$
We take the partition $t_0=0<t_1<t_2<...<t_{N-1}<t_N=T$ of $[0,T]$, where $h:=T/N$ and $t_j=jh$ for $j=0,1,2,...N.$ By definition of $y'$, we will have $y'(t)\approx\frac{y(t+h)-y(t)}{h}$, for $h$ small enough. So, if $N$ is large enough, we will infer $f(y(t))\approx\frac{y(t+h)-y(t)}{h}$ or $y(t+h)\approx y(t)+hf(y(t)).$ Hence, if we take $t=t_j$, then
$y(t_{j+1})\approx y(t_j)+hf(y(t_j)).$ Thus, we can define a set $\{y_j\}_{j=0}^N$ where, $y_{j+1}=y_j+hf(y_j)$ and $y_j\approx y(t_j)$.

our case says $P_{0}=5$ and $T=20.$ Also, $h=0.5,\ 0.25$ and $h=0.1$ (or equivalently)

The exact solution is $P(t)=5e^{0.25t}$ and the plot can be seen in figure below:

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# The last two social security number, A = 74 
epsSTUD = 15 * 1e-2 


# TASK 1

## Part a

p0 = 5                               ## Initial number of parasite
k = 0.1 + epsSTUD                    ## Population growth rate
tExact = np.arange(0, 20.05, 0.05)   ## Time interval (set to > 20 so 
                                     ## 20 is included) 
pExact = 5 * math.e ** (k * tExact)  ## Exact solutionas a function of tExact


#Plot the exact solution
plt.figure()
plt.plot(tExact, pExact, label='Data')
plt.xlabel('time')
plt.ylabel('parasite count')
plt.legend()
plt.show()
```

We will have some approximate solutions in a consequence of the forward Euler method
with respect to values of N ’s. The plots for different values of h can be seen in figure blow. As the value of h decreases, the numerical solution gets closer to the exact solutio

```python
## Part b

### h = 0.5

T = 20
h0 = 0.5                               ## Step size
N0 = int(T / h0)                       ## Number of steps used by Euler's 
                                       ## method
tEuler0 = np.zeros((N0+1))
pEuler0 = np.zeros((N0+1))

## Initial time and initial value
tEuler0[0] = 0
pEuler0[0] = p0

for i in range (1, N0+1):
    tEuler0[i] = tEuler0[i-1] + h0
    pEuler0[i] = pEuler0[i-1] + h0 * (k * pEuler0[i-1])
    
    
### h = 0.25

T = 20
h1 = 0.25                              ## Step size
N1 = int(T / h1)                       ## Number of steps used by Euler's 
                                       ## method
tEuler1 = np.zeros((N1+1))
pEuler1 = np.zeros((N1+1))

## Initial time and initial value
tEuler1[0] = 0
pEuler1[0] = p0

for i in range (1, N1+1):
    tEuler1[i] = tEuler1[i-1] + h1
    pEuler1[i] = pEuler1[i-1] + h1 * (k * pEuler1[i-1])


### h = 0.1

T = 20
h2 = 0.1                               ## Step size
N2 = int(T / h2)                       ## Number of steps used by Euler's 
                                       ## method
tEuler2 = np.zeros((N2+1))
pEuler2 = np.zeros((N2+1))

## Initial time and initial value
tEuler2[0] = 0
pEuler2[0] = p0

for i in range (1, N2+1):
    tEuler2[i] = tEuler2[i-1] + h2
    pEuler2[i] = pEuler2[i-1] + h2 * (k * pEuler2[i-1])    

# Plot the numerical solution
plt.figure()
plt.plot(tExact, pExact, label='Exact solution')
plt.plot(tEuler0, pEuler0, label='h = 0.5')
plt.plot(tEuler1, pEuler1, label='h = 0.25')
plt.plot(tEuler2, pEuler2, label='h = 0.1')
plt.xlabel('time')
plt.ylabel('parasite count')
plt.legend()
plt.show()
```

The loglog command is used for creating a log-log plot, meaning both the x-axis and y-axis are displayed on a logarithmic scale. This choice is often made when dealing with error analysis or when the range of values spans several orders of magnitude.

In the specific context of error analysis, using a log-log plot is beneficial because it helps visualize the convergence rate of the numerical method, such as Euler's method, in a clearer way. The reference line of slope 1 represents a linear relationship on a log-log scale, corresponding to an error that decreases at a rate proportional to the step size. In other words, if the error is proportional to the step size raised to the power of 1, it will form a straight line with a slope of 1 when plotted on a log-log scale.

When we observe a line with a slope close to 1 on a log-log plot, it indicates that the numerical method is converging at a rate consistent with the expected behavior. Deviations from this straight line might suggest different convergence behavior or errors in the numerical implementation.

we have $H_0=500,\ Z_0=10,\ R_0=0,\ \alpha=0.00025,\ \beta=0.01,\ \zeta=0.02$ and $h=0.1$.

The results for two different step sizes $h = 0.65$ and $h = 0.1$ can be seen in figures. Step size $h = 0.65$ produces an unstable or fluctuating result, while step size $h = 0.1$ produces a stable result. 

```python
# TASK 2

h = 0.65                               ## Step size
T_end = 10                             ## Time
N = int(T_end / h)                     ## Number of steps used by Euler's 
                                       ## method 

## Allocate arrays
t   = np.zeros(N + 1)                  ## Time
zom = np.zeros(N + 1)                  ## Zombie population
hum = np.zeros(N + 1)                  ## Human population
rem = np.zeros(N + 1)                  ## Removed zombie population
tot = np.zeros(N + 1)                  ## Total population

## Initial conditions    
t[0]   = 0
zom[0] = 10
hum[0] = 500
rem[0] = 0 

# Constants
alpha = 0.005 * epsSTUD
beta  = 0.01
zeta  = 0.02

## Calculate the population of humans, zombies, and removed zombies
for i in range (1, N+1):
    t[i] = t[i-1] + h0
    hum[i] = hum[i-1] + h * (- beta * hum[i-1] * zom[i-1])
    zom[i] = zom[i-1] + h * (beta * hum[i-1] * zom[i-1] + zeta * rem[i-1]
                              - alpha * hum[i-1] * zom[i-1])
    rem[i] = rem[i-1] + h * (alpha * hum[i-1] * zom[i-1] - zeta * rem[i-1]) 
    
## Calculate the total population
tot = hum + zom + rem


## Plot the numerical solution
plt.figure()
plt.plot(t, hum, label='Human population')
plt.plot(t, zom, label='Zombie population')
plt.plot(t, rem, label='Removed zombie population')
plt.plot(t, tot, label='Total population')
plt.xlabel('time')
plt.ylabel('Number of individual')
plt.legend()
plt.show()
```
