Introduction:

Problem

$$u_{t}(x,t) - u_{xx}(x,t) = \frac{B}{2} (2t + 1 - x^2) , 0<x<1 , 0<t<T

u_x(0,t) = u(1,t) = 0 , 0<t<T

u(x,0) = u_0(x) = 0 , 0<x<1.$$

The exact solution: 

$$u(x,t) = - \frac{B}{2} (tx^2 - t) = \frac{B}{2} (1 - x^2)t,$$

where, $B:={\rm epsSTUD}$.
The variational formulation to get a FEM:

$$ (1) \int_{0}^1 u_{h,t} (x,t) v_h (x) dx + \int_0^1 u_{h,x} (x,t) v'_h (x)dx = \int_0^1 f(x,t) v_h (x) dx , \ \forall v_h \in V_h  $$

where, 
$$ f(x,t) = \frac{B}{2} (2t + 1 - x^2), \ (x,t) \in (0,1) \times (0,T) $$

and $ V_h $ is a subspace of
$$ V = \{ w: [0,1] \to \mathbb{R} | w \in H^1 (0,1) \ \ \& \ \ w(1) = 0 \} $$

consisting of continuous functions that are piecewise linear on the partition ${\cal T}_h$ that 

$ {\cal T}_{h} = jh \ such \ that \  j = 0, 1, ... , m+1 \ , \   h = \frac{1}{m+1} $.

Let $\{ \phi_j \}_{j=0}^m$ be the standard basis of $ V_h $ consisting of the usual hat functions. So, we can write

$$ (2)  u_h (x,t) = \sum_{j=0}^m \xi_j (t) \phi_j (x).$$

By inserting $(2)$ in $(1)$ and choosing as test functions $ v_h = \phi_i $ \ \ for \ \ $i = 0, 1, ..., m$, we then obtain 
$$ M \xi' (t) + S \xi (t) = F(t) $$

with $\xi_j (0) = 0$ \ \ for \ \ $ j = 0, 1, ..., m$ , where,  $ M= (m_{ij}), \ \ S=(s_{ij})$ \ \ and \ \  $ F(t) = (F_i (t))$ which are

$$ m_{ij} = \int_0^1 \phi_i (x) \phi_j (x) dx ,\ \ s_{ij} = \int_0^1 \phi'_i (x) \phi'_j (x) dx, \ \ F_i (t) = \int_0^1 f(x,t) \phi_i (x) dx , \ \ i,j = 0, 1, ..., m .$$

By calculating: $ m_{i,i+1} = \frac{h}{6} $ for $ i = 0, ..., m-1 $, and $ m_{ii} = \frac{2h}{3} $ for $ i = 1, ..., m $, and $ m_{00} = \frac{h}{3} $, and $ m_{ij} = 0 $ for $ |i-j|>1 $, and $ s_{i,i+1} = \frac{-1}{h} $ for $ i = 0, ..., m-1 $, and $ s_{ii} = \frac{2}{h} $ for $ i = 1, ..., m $, and $ s_{00} = \frac{1}{h} $, and $ s_{ij} = 0 $ for $ |i-j|>1 $, and $ F_0(t) = \frac{hB}{24} (6(2t + 1) - h^2) $, $ F_j(t) = \frac{hB}{12} (6(2t + 1) - (6j^2 + 1) h^2) $ for $ j=1, ..., m $.

Thus,
$$M_{(m+1) \times (m+1)} =\frac{h}{6}\left( {\begin{array}{cccccc}
    2 & 1 & 0 & 0 & \cdots & 0 \\
    1 & 4 & 1 & 0 & \cdots & 0 \\
    0 & 1 & 4 & 1 & \cdots & 0 \\
    0 & 0 & 1 & 4 & \cdots & 0 \\
    \vdots & \vdots & \vdots & \vdots & \ddots &  1\\
    0 & 0 & \cdots & 0 & 1 & 4 \\
  \end{array} } \right)
$$
and, 
$$S_{(m+1) \times (m+1)} =\frac{1}{h}\left( {\begin{array}{cccccc}
    1 & -1 & 0 & 0 & \cdots & 0 \\
    -1 & 2 & -1 & 0 & \cdots & 0 \\
    0 & -1 & 2 & -1 & \cdots & 0 \\
    0 & 0 & -1 & 2 & \cdots & 0 \\
    \vdots & \vdots & \vdots & \vdots & \ddots & -1\\
    0 & 0 & \cdots & 0 & -1 & 2 \\
  \end{array} } \right)
$$
and, 
$$F(t)_{(m+1) \times 1} =\left( {\begin{array}{cccc}
    F_0(t) \\
    F_1(t) \\
    \vdots \\
    F_m(t) \\
  \end{array} } \right)
$$
and, 
$$\xi(t)_{(m+1) \times 1} =\left( {\begin{array}{cccc}
    \xi_0(t) \\
    \xi_1(t) \\
    \vdots \\
    \xi_m(t) \\
  \end{array} } \right)
$$
and 
$$ M \xi' (t) + S \xi (t) = F(t) , \ \ \xi_j(0) = 0 , \ \ j=0, 1, ..., m $$

Result:

The last two digits of the student number used were 74, thus $epsSTUD = 0.074$. The computation was done in the time range $0<t<T$, where $T = 10$ s and time interval $k = 1 s$. The spatial domain was divided into 11 nodes with $h = 0.1$. The integration of function $f(x,t)$ and $v_h$ was approximated using midpoint rule.

The results are depicted in figure \ref{fig:sol} below. As can be seen, the numerical result obtained was in agreement with the exact solution in all timestep.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

epsSTUD = 74 * 10**(-3)

# Domain
a     = 0                     ## domain [0,1]
b     = 1       


# Exact solution
n_exact = 100
x_exact = np.linspace(a, b, n_exact)

def u_exact(x, t):
    return -(epsSTUD / 2) * (t*x**2 - t)

# Mesh generation
m           = 9                        ## number of internal nodes
mt          = m + 2                    ## number of total nodes
h           = (b-a) / (m+1)            ## mesh size
T           = 10                       ## time T
n           = 10                       ## timestep / subinterval
k           = T/n


# Array allocation
M           = np.zeros((mt,3))         ## mass matrix (index 0 for i-1, 1 for 
                                       ## i, and 2 for i+1
S           = np.zeros((mt,3))         ## stiffness matrix
A           = np.zeros((mt,3))         ## M + S
F           = np.zeros((mt))           ## load vector
xi          = np.zeros((mt, n+1))      ## xi
u_h         = np.zeros((mt, n+1))      ## approximate solution
rhs         = np.zeros((mt))           ## right-hand side, to temporary store
                                       ## (M xi(l-1) + k F(tl)) 
U           = np.zeros((n_exact, n+1)) ## exact solution

# Mesh generation
x     = np.linspace(a, b, mt)

# Time partition
t     = np.linspace(0, T, n + 1)  ## add 1 since t_0 = 0 is included



# Midpoint rule for load vector

## the first term of the integral
def f1(x1, x2, t):
    # int_a = (-x1**2 + 2*t + 1) * (x1 - x1)    
    # int_b = (-x2**2 + 2*t + 1) * (x2 - x1)
    
    # int_tot = (h / 2) * (int_b - int_a)
    
    mid = (x1 + x2) / 2
    int_tot = (x2 - x1) * (-mid**2 + 2*t + 1) * (mid - x1)
    
    return int_tot

## the second term of the integral
def f2(x2, x3, t):
    # int_b = (-x2**2 + 2*t + 1) * (x3 - x2)
    # int_c = (-x3**2 + 2*t + 1) * (x3 - x3)
    
    # int_tot = (h / 2) * (int_c - int_b)
    
    mid = (x3 + x2) /2
    int_tot = (x3 - x2) * (-mid**2 + 2*t + 1) * (x3 - mid)
    
    return int_tot


# Calculate the mass and sitffness matrix, and load vector

## mass and stiffness matrix (for the internal nodes)
for i in range (1, mt-1):
    ### M[i,i-1]
    M[i,0] = (x[i]**3 - 3*(x[i]**2)*(x[i-1]) + 3*(x[i])*(x[i-1]**2) 
                  - x[i-1]**3) / (6 * h**2)
    ### M[i,i]
    M[i,1] = ((-3*(x[i-1])*(x[i]**2) + 3*(x[i]**2)*(x[i+1]) 
              + 3*(x[i-1]**2)*(x[i]) - 3*(x[i])*(x[i+1]**2) 
              - x[i-1]**3 + x[i+1]**3) / (3 * h**2))
    ### M[i,i+1]
    M[i,2] = (x[i+1]**3 - 3*(x[i+1]**2)*(x[i]) + 3*(x[i+1])*(x[i]**2) 
              - x[i]**3) / (6 * h**2)
    
    ### S[i,i-1]
    S[i,0] = -1/h
    ### S[i,i]
    S[i,1] =  2/h
    ### S[i,i+1]
    S[i,2] = -1/h

    
## mass and stiffness matrix (for node at x = 0)

### M[0,0]
M[0,1] = (x[i+1] - x[i])**3 / (3 * h**2)
### M[0,1]
M[0,2] = (x[1]**3 - 3*(x[1]**2)*(x[0]) + 3*(x[1])*(x[0]**2) 
          - x[0]**3) / (6 * h**2)

### S[0,0]
S[0,1] =  1/h
### S[0,1]
S[0,2] = -1/h
    
## matrix A
A = M + k * S



# Initial condition
xi[:,0] = 0


# Iteration (index i is for space and index l is for time)
for l in range (1,n+1):
    
    ## compute the load vector (for inner node)
    for i in range (0, mt-1):
        if i == 0:
            F[0] = (0.037 / h) * (f2(x[0], x[1], t[l]))        
        else:
            F[i] = (0.037 / h) * (f1(x[i-1], x[i], t[l]) 
                                  + f2(x[i], x[i+1], t[l]))
           

    ## compute the right-hand side (M xi(l-1) + k F(tl)):
    for i in range (0, mt-1):
        if i == 0:
            rhs[i] = (M[i,1] * xi[i,l-1] + M[i,2] * xi[i+1,l-1]) + (k * F[i])
        else:
            rhs[i] = (M[i,0] * xi[i-1,l-1] + M[i,1] * xi[i,l-1] 
                      + M[i,2] * xi[i+1,l-1]) + (k * F[i])
    
    ## compute xi using Gauss-Seidel method
    for iter in range (1000):        
        for i in range (0, mt-1):
            if i == 0:
                xi[i,l] = (rhs[i] - A[i,2] * xi[i+1,l]) / A[i,1]
            else:
                xi[i,l] = (rhs[i] - A[i,0] * xi[i-1,l] 
                           - A[i,2] * xi[i+1,l]) / A[i,1]
   
    u_h = xi
                
    ## compute exact solution
    U[:,l] = u_exact(x_exact, t[l]) 
    
    ## Plot the exact and numerical solution in each timestep
    plt.figure()
    plt.plot(x_exact, U[:,l], label='exact solution')
    plt.plot(x, u_h[:,l], label='numerical solution')
    plt.xlabel('mesh')
    plt.ylabel('u')
    plt.title(f"t = {t[l]} s")
    plt.legend()
    plt.show()
    
```
